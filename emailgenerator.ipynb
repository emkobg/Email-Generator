```python
#defined the get_punct fucntion, that takes email prefix and returns the punctuiation is one is found
def get_punct(email_prefix):
    punct_return = None
    
    for punct in punctuation:
        if punct in email_prefix:
            punct_return = punct
            break
        
    return punct_return
```

```python
#testing if the function works 
email_prefix = 'emil.bobev'

get_punct(email_prefix)
```

```python
# find all possible patterns
def pattern_match(first, last, email):
    first_name = first.lower()
    last_name = last.lower() 
    
    if type(email) == float: 
        return 'Email is Empty'
    
    email_prefix = email.split('@')[0].lower()
    domain = email.split('@')[1].lower()
    punct = get_punct(email_prefix)
    
    

    if punct:
        if first_name == email_prefix[:len(first)] and last_name == email_prefix[-len(last):]:
            return {'pattern_match': 'first_name_last_name','punct': punct, 'domain': domain}
        elif first_name != email_prefix[:len(first)] and last_name == email_prefix[-len(last):] and first_name[0] == email_prefix.split(punct)[0]:
            return {'pattern_match': 'first_initial_last_name','punct': punct, 'domain': domain}
        elif first_name == email_prefix[:len(first)] and last_name != email_prefix[-len(last):] and last_name[0] == email_prefix.split(punct)[1]:
            return {'pattern_match': 'first_name_last_initial','punct': punct, 'domain': domain}
        elif last_name == email_prefix[:len(last)] and first_name[0] == email_prefix.split(punct)[1]:
            return {'pattern_match': 'last_name_first_initial','punct': punct, 'domain': domain}
        elif last_name[0] == email_prefix.split(punct)[0]  and first_name == email_prefix.split(punct)[1]:
            return {'pattern_match': 'last_initial_first_name','punct': punct, 'domain': domain}
        elif last_name == email_prefix.split(punct)[0]  and first_name == email_prefix.split(punct)[1]:
            return {'pattern_match': 'last_name_first_name','punct': punct, 'domain': domain}
        
        else:
            return 'Email given, pattern not found'
    
    else: 
        if first_name == email_prefix[:len(first)] and last_name == email_prefix[-len(last):]:
            return {'pattern_match': 'first_name_last_name','punct': punct, 'domain': domain}
        elif first_name[0] == email_prefix[0] and last_name == email_prefix[-len(last):] and len(last_name) + 1 == len(email_prefix):
            return {'pattern_match': 'first_initial_last_name','punct': punct, 'domain': domain}
        elif first_name == email_prefix[:len(first)] and last_name[0] == email_prefix[-1] and len(first_name) + 1 == len(email_prefix):
            return {'pattern_match': 'first_name_last_initial','punct': punct, 'domain': domain}
        elif last_name == email_prefix[:len(last)] and first_name[0] == email_prefix[-1] and len(last_name) +1 == len(email_prefix):
            return {'pattern_match': 'last_name_first_initial','punct': punct, 'domain': domain}
        elif last_name[0] == email_prefix[0]  and first_name == email_prefix[1:]:
            return {'pattern_match': 'last_initial_first_name','punct': punct, 'domain': domain}
        elif last_name == email_prefix[:len(last)]  and first_name == email_prefix[-len(first):] and len(email_prefix) == len(first_name)+ len(last_name):
            return {'pattern_match': 'last_name_first_name','punct': punct, 'domain': domain}
        elif email_prefix == first_name :
            return {'pattern_match': 'first_name','punct': punct, 'domain': domain}
        elif email_prefix == last_name:
            return {'pattern_match': 'last_name','punct': punct, 'domain': domain}
        
        else:
            return 'Email given, pattern not found'
        

pattern_match ('Emil','Bobev', 'bobev@stockton.edu')

```

```python
#create a dataframe email2 which creates a pattern_match column
email['pattern_match'] = email[["First Name","Last Name", "Email"]].apply(lambda x: pattern_match(*x), axis=1)
email
```

```python
#modifiing the dataframe so it breaks it down and displays patter, punctuation and checker(how the email is created - pattern + url)
email['pattern'] = email.apply(lambda _: '',axis=1)
email['punctuation'] = email.apply(lambda _: '',axis=1)
email['domain'] = email.apply(lambda _: '',axis=1)
email['checker'] = email.apply(lambda _: '',axis=1)

for index, row in email.iterrows():
    pattern = row['pattern_match']
    if type(pattern) == dict: 
        row['pattern']=pattern.get('pattern_match')
        row['punctuation']=pattern.get('punct')
        row['domain']=pattern.get('domain')
        row['checker']= pattern.get('pattern_match') + str(pattern.get('punct')) + pattern.get('domain')
    else: 
        row['pattern']='pass'
        row['punctuation']='pass'
        row['domain']='pass'
        row['checker'] = 'pass'
        
email
        
```

```python
#creating a dataframe only for existing emails
email2 = email[email['checker'] != 'pass']
email2
```

```python
#verifiying if the companies have their email with the checker 
valid = email2.groupby('Company Name')['checker'].nunique() ==1 
valid
```

```python
def valid_company(string): 
    return valid[f'{string}']
```

```python
# checking if there is an email provided or the email provided matches any of the patterns
def usable_company(company, pattern): 
    if pattern == 'Email is Empty': 
        return False
    elif pattern == 'Email given, pattern not found': 
        return False 
    else :
        return valid_company(f'{company}')
```
